# An Introduction to Plotting

Matplotlib is the common package used in python.  It is very easy to use for simple plots, but it can also be coerced into making complicated plots and graphics for publication.  

We will start with simple functional plots.

## Initiating numpy and matplotlib

I often put all of the stuff I need to load at the top of a notebook, although it is not necessary.  

The first line in the following cell creates plots that are embedded in the notebook.  

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

## Size of the plot

The following command will create a single nice large plot.

```python
plt.rcParams["figure.figsize"] = (10,7)
```

## Create some data using numpy

Can you describe in words what the vector X will be?

```python
X = np.linspace(-np.pi, np.pi, 256, endpoint=True)
C, S = np.cos(X), np.sin(X)
```

## A simple plot is easy

```python
plt.plot(X,C)
plt.plot(X,S)
```

## Customizing the plot

This is where the real artistry happens.  In the next slides, we will look at how to adjust various feature of the plot.

### Adjust the size and line properties

```python
plt.figure(figsize=(10,6), dpi=80)
plt.plot(X, C, color="blue", linewidth=2.5, linestyle="-")
plt.plot(X, S, color="red",  linewidth=2.5, linestyle="-")
```

### Colors

Matplotlib colors can be found here:
https://matplotlib.org/3.1.0/gallery/color/named_colors.html

```python
plt.figure(figsize=(10,6), dpi=80)
plt.plot(X, C, color="lightseagreen", linewidth=2.5, linestyle="-")
plt.plot(X, S, color="darksalmon",  linewidth=2.5, linestyle="-")
```

###  Setting  limits

i.e., the x and y ranges for the plot.

```python
# same stuff
plt.figure(figsize=(10,6), dpi=80)
plt.plot(X, C, color="blue", linewidth=2.5, linestyle="-")
plt.plot(X, S, color="red",  linewidth=2.5, linestyle="-")

# new stuff
plt.xlim(X.min()*1.2, X.max()*1.2)
plt.ylim(C.min()*1.2, C.max()*1.2)
```

### Setting ticks

Current ticks are not ideal because they do not show the interesting values (+/-,+/-/2) for sine and cosine. We'll change them such that they show only these values.

```python
# same stuff
plt.figure(figsize=(10,6), dpi=80)
plt.plot(X, C, color="blue", linewidth=2.5, linestyle="-")
plt.plot(X, S, color="red",  linewidth=2.5, linestyle="-")
plt.xlim(X.min()*1.2, X.max()*1.2)
plt.ylim(C.min()*1.2, C.max()*1.2)

#new stuff
plt.xticks( [-np.pi, -np.pi/2, 0, np.pi/2, np.pi])
plt.yticks([-1, 0, +1])
```

Ticks are now properly placed but their label is not very explicit. We could guess that 3.142 is  but it would be better to make it explicit. When we set tick values, we can also provide a corresponding label in the second argument list. 

Note that latex is used for nice rendering of the label.

```python
# same stuff
plt.figure(figsize=(10,6), dpi=80)
plt.plot(X, C, color="blue", linewidth=2.5, linestyle="-")
plt.plot(X, S, color="red",  linewidth=2.5, linestyle="-")
plt.xlim(X.min()*1.2, X.max()*1.2)
plt.ylim(C.min()*1.2, C.max()*1.2)

# different stuff

plt.xticks([-np.pi, -np.pi/2, 0, np.pi/2, np.pi],
       [r'$-\pi$', r'$-\pi/2$', r'$0$', r'$+\pi/2$', r'$+\pi$'])
plt.yticks([-1, 0, +1],
       [r'$-1$', r'$0$', r'$+1$'])
```

### Adding a legend

Legends are very useful.  The information about what the red curve corresponds to could be given in the figure caption, but it is easier to understand and appreciate at a glance if given in the figure.

More info about legends is here:
https://matplotlib.org/tutorials/intermediate/legend_guide.html

```python
# same stuff

plt.figure(figsize=(10,6), dpi=80)
plt.plot(X, C, color="blue", linewidth=2.5, linestyle="-")
plt.plot(X, S, color="red",  linewidth=2.5, linestyle="-")
plt.xlim(X.min()*1.2, X.max()*1.2)
plt.ylim(C.min()*1.2, C.max()*1.2)
plt.xticks([-np.pi, -np.pi/2, 0, np.pi/2, np.pi],
       [r'$-\pi$', r'$-\pi/2$', r'$0$', r'$+\pi/2$', r'$+\pi$'])
plt.yticks([-1, 0, +1],
       [r'$-1$', r'$0$', r'$+1$'])

# New stuff

plt.plot(X, C, color="blue", linewidth=2.5, linestyle="-", label="cosine")
plt.plot(X, S, color="red", linewidth=2.5, linestyle="-", label="sine")

plt.legend(loc='upper left', frameon=False)
```

### Fixing it up

This plot is not publication ready.  There are no axes labels, and the tick labels are too small.  

The axes class (collection of functions) can take many arguments, allowing for fine control of many aspects of the figure's appearane.  All the commands can be found here:

https://matplotlib.org/3.1.1/api/axes_api.html#the-axes-class

```python
# different stuff:

fig, ax = plt.subplots(figsize=(10,6), dpi=80)
ax

# same stuff
plt.xlim(X.min()*1.2, X.max()*1.2)
plt.ylim(C.min()*1.2, C.max()*1.2)
plt.plot(X, C, color="blue", linewidth=2.5, linestyle="-", label="cosine")
plt.plot(X, S, color="red", linewidth=2.5, linestyle="-", label="sine")

# different and new stuff

plt.xticks([-np.pi, -np.pi/2, 0, np.pi/2, np.pi],
       [r'$-\pi$', r'$-\pi/2$', r'$0$', r'$+\pi/2$', r'$+\pi$'],fontsize=18)

plt.yticks([-1, 0, +1],
       [r'$-1$', r'$0$', r'$+1$'],fontsize=18)

plt.legend(loc='upper left', frameon=True,fontsize=18)

ax.tick_params('both',direction='in',length=8, width=2)
ax.set_xlabel('Phase',fontsize=22)
ax.set_ylabel('Amplitude',fontsize=22)
ax.spines['left'].set_linewidth(2)
ax.spines['right'].set_linewidth(2)
ax.spines['top'].set_linewidth(2)
ax.spines['bottom'].set_linewidth(2)
```
