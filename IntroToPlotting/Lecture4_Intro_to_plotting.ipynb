### Setting ticks

The current ticks are not ideal because they do not show the interesting values (+/-π,+/-π/2) for sine and cosine. We'll change them such that they show only these values.

````python
# same stuff
plt.figure(figsize=(10,6), dpi=80)
plt.plot(X, C, color="blue", linewidth=2.5, linestyle="-")
plt.plot(X, S, color="red",  linewidth=2.5, linestyle="-")
plt.xlim(X.min()*1.2, X.max()*1.2)
plt.ylim(C.min()*1.2, C.max()*1.2)

#new stuff
plt.xticks( [-np.pi, -np.pi/2, 0, np.pi/2, np.pi])
plt.yticks([-1, 0, +1])
````

Ticks are now properly placed but their label is not very explicit. We could guess that 3.142 is π but it would be better to make it explicit. When we set tick values, we can also provide a corresponding label in the second argument list. 

Note that latex is used for nice rendering of the label.

````python
# same stuff
plt.figure(figsize=(10,6), dpi=80)
plt.plot(X, C, color="blue", linewidth=2.5, linestyle="-")
plt.plot(X, S, color="red",  linewidth=2.5, linestyle="-")
plt.xlim(X.min()*1.2, X.max()*1.2)
plt.ylim(C.min()*1.2, C.max()*1.2)

# different stuff

plt.xticks([-np.pi, -np.pi/2, 0, np.pi/2, np.pi],
       [r'$-\pi$', r'$-\pi/2$', r'$0$', r'$+\pi/2$', r'$+\pi$'])
plt.yticks([-1, 0, +1],
       [r'$-1$', r'$0$', r'$+1$'])
````

### Adding a legend

Legends are very useful.  The information about what the red curve corresponds to could be given in the figure caption, but it is easier to understand and appreciate at a glance if given in the figure.

More info about legends is here:
https://matplotlib.org/tutorials/intermediate/legend_guide.html

````python
# same stuff

plt.figure(figsize=(10,6), dpi=80)
plt.plot(X, C, color="blue", linewidth=2.5, linestyle="-")
plt.plot(X, S, color="red",  linewidth=2.5, linestyle="-")
plt.xlim(X.min()*1.2, X.max()*1.2)
plt.ylim(C.min()*1.2, C.max()*1.2)
plt.xticks([-np.pi, -np.pi/2, 0, np.pi/2, np.pi],
       [r'$-\pi$', r'$-\pi/2$', r'$0$', r'$+\pi/2$', r'$+\pi$'])
plt.yticks([-1, 0, +1],
       [r'$-1$', r'$0$', r'$+1$'])

# New stuff

plt.plot(X, C, color="blue", linewidth=2.5, linestyle="-", label="cosine")
plt.plot(X, S, color="red", linewidth=2.5, linestyle="-", label="sine")

plt.legend(loc='upper left', frameon=False)
````

### Fixing it up

This plot is not publication ready.  There are no axes labels, and the tick labels are too small.  

The axes class (collection of functions) can take many arguments, allowing for fine control of many aspects of the figure's appearane.  All the commands can be found here:

https://matplotlib.org/3.1.1/api/axes_api.html#the-axes-class

````python
# different stuff - define the ax in order to alter it.

fig, ax = plt.subplots(figsize=(10,6), dpi=80)

# same stuff
plt.xlim(X.min()*1.2, X.max()*1.2)
plt.ylim(C.min()*1.2, C.max()*1.2)
plt.plot(X, C, color="blue", linewidth=2.5, linestyle="-", label="cosine")
plt.plot(X, S, color="red", linewidth=2.5, linestyle="-", label="sine")

# different and new stuff

# making the tick labels bigger:

plt.xticks([-np.pi, -np.pi/2, 0, np.pi/2, np.pi],
       [r'$-\pi$', r'$-\pi/2$', r'$0$', r'$+\pi/2$', r'$+\pi$'],fontsize=18)

plt.yticks([-1, 0, +1],
       [r'$-1$', r'$0$', r'$+1$'],fontsize=18)

# making a legend

plt.legend(loc='upper left', frameon=True,fontsize=18)

# we want ticks on the top and right axes.  We also want them pointing inward.

ax.tick_params(top=True,right=True,direction='in',length=8, width=2)

# set the labels for the axes. Make them big enough.

ax.set_xlabel('Phase (Radians)',fontsize=22)
ax.set_ylabel('Amplitude ($10^8$ cm)',fontsize=22)

# make the axes thicker

ax.spines['left'].set_linewidth(2)
ax.spines['right'].set_linewidth(2)
ax.spines['top'].set_linewidth(2)
ax.spines['bottom'].set_linewidth(2)
````

### Try this

One of the common functions in astronomy is a power law.  For example, the optical spectrum of a quasar can have the following form:

$$F(\lambda)=\Big(\frac{\lambda}{\lambda_0}\Big)^\alpha$$

where $\lambda$ is the wavelength in Angstroms, $\lambda_0$ is some convenient reference wavelength in the bandpass of interest, and $\alpha$ is the power law slope.

- Create a wavelength independent variable running between 4000 and 8000 Angstroms and having 4000 points.
- Create two flux-density dependent variables, call them $F_1$ and $F_2$, which have power law indices of $\alpha=-1.85$ and $\alpha=0.1$, respectively.  Label the first one "Blue Quasar" and the second one "Red Quasar".  The units of flux density are $\rm 10^{-17}\,erg\,s^{-1}\,cm^{-2}\, A^{-1}$.  Write the function so that the value at 6000 angstroms is 1.0. 
- Make a plot showing these two functions.  Choose clearly contrasting colors.  Label the x-axis "Wavelength (Angstroms)" and the y-axis in flux density units.  Rendering the angstrom symbol is problematic, so you can just use "A".  Also, plot the y-axis from zero flux density.
- Adjust sizes of fonts, etc., to make everything look nice.

Hints:
- Don't try to make the final plot all at once.  Do it as we have done in the example - make the functions, then make a rough plot as a sanity check to make sure you've done the functions right, and then make the plot nice.

````python
wavelength=np.linspace(4000, 8000, 4000, endpoint=True)
wavelength0=6000.0
f1=(wavelength/wavelength0)**(-1.85)
f2=(wavelength/wavelength0)**(0.1)
plt.plot(wavelength,f1)
plt.plot(wavelength,f2)
````

````python
fig, ax = plt.subplots(figsize=(10,6), dpi=80)


plt.plot(wavelength,f1,'b',linewidth=2.5, linestyle="-", label="blue quasar")
plt.plot(wavelength,f2,'r',linewidth=2.5, linestyle="-", label="red quasar")

plt.xlim(wavelength.min()-250, wavelength.max()+250)
plt.ylim(0, f1.max()*1.1)

plt.legend(loc='upper right', frameon=True,fontsize=15)

ax.set_xlabel('Wavelength (Angstroms)',fontsize=16)
ax.set_ylabel('Flux Density ($10^{-17}$ erg s$^{-1}$ cm$^{-2}$ A$^{-1}$)',fontsize=16)

# make the axes thicker

ax.spines['left'].set_linewidth(2)
ax.spines['right'].set_linewidth(2)
ax.spines['top'].set_linewidth(2)
ax.spines['bottom'].set_linewidth(2)

plt.xticks(fontsize=15)
plt.yticks(fontsize=15)

ax.tick_params(top=True,right=True,direction='in',length=8, width=2)

````
